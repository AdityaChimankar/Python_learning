# IF statement

```python
# in the earlier exercises, we focused on evaluating boolean expressions as True or False.
# the IF statement uses this to decide the next lines to execute
```

```python
marks = 761

if (marks >= 40) and (marks < 60):
    print("Pass")
elif (marks >= 60) and (marks < 75):
    print("First class")
elif (marks >= 75) and (marks <= 100):
    print("Distinction")
elif (marks < 40):
    print("Try again")
else:
    print("Incorrect marks entered")
```

```python
marks = 78
if (marks >= 40):
    print("Pass")
```

```python
marks = 78
result = ""

if (marks < 40):
    result = "Fail"
else:
    result = "True"

print(result)
```

```python
# there is one more way of writing this

marks = 78
result = ""

if (marks < 40):
    result = "Fail"

if (marks >= 40):
    result = "True"
print(result)
```

```python
# Exercise:
# Students are eligible for taking exams if their attendance is at least 75%
# take the total number of classes conducted and the attendance of a student
# if less than cut-off but the student has an exception appoved by the Principal then they are allowed.
# Else permission is denied
```

```python
# calculate attendance percentage
attended = 40
classes_conducted = 55
attendance = attended/classes_conducted * 100

# is exception approved?
exception_request = "Approved"

# check if attendance is 75% or more
if attendance >= 75:
    print("eligible for exams")
else:
    # eligibility only based on exception approval
    if (exception_request == "Approved"):
        print("eligible based on exception approval")
    else:
        # student is not eligible
        print("not eligible")


```

```python
# check if attendance is 75% or more
if (attendance >= 75) or (exception_request == "Approved"):
        print("eligible")
else:
    # student is not eligible
    print("not eligible")
```
